```python
edges = [20,30,40,50,60,70,80,90,100,110,120,130,140,150,160,170,180]
step = 10
```

```python
labels=('prediction', 'ground truth')
columns = np.array(edges[:-1]) + step / 2
width = 0.3 * step
fig, ax1 = plt.subplots(figsize=(12, 8))
p1 = ax1.bar(columns - width / 2, pred, width, color='b', label=labels[0]+' counts')
p2 = ax1.bar(columns + width / 2, ground, width, color='g', label=labels[1]+' counts')
plt.xticks(edges, rotation=30)
ax1.set_xlabel('Brightness (average of intensity)')
ax1.set_ylabel('Number of loops')
ax2 = ax1.twinx()
p_values = precision
r_values = recall
p3 = ax2.plot(columns, p_values, 'ro--', label='precision')
p4 = ax2.plot(columns, r_values, 'co--', label='recall')
plt.ylim(0,1)
ax2.set_ylabel('Score')
lines, labels = ax1.get_legend_handles_labels()
lines2, labels2 = ax2.get_legend_handles_labels()
ax2.legend(lines + lines2, labels + labels2, bbox_to_anchor=(1.04, 1), loc="upper left")
```
